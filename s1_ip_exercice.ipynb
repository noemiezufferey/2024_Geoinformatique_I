```python
# Ce script permet de charger une couche géospatiale d'une zone industrielle depuis un fichier Geopackage (.gpkg),
# puis de générer des zones tampons (buffers) autour de cette zone à différentes distances (500m, 1km, et 5km).
# Ensuite, il calcule les anneaux de ces buffers, c'est-à-dire les zones situées entre ces distances (par exemple,
# entre 500m et 1km, et entre 1km et 5km).
# Les zones tampons et anneaux sont ensuite fusionnées, sauvegardées dans un fichier GeoJSON, et visualisées sur
# une carte avec la zone industrielle, les tampons et les anneaux affichés en différentes couleurs.

import geopandas as gpd
import pyogrio
from shapely.ops import unary_union
import matplotlib.pyplot as plt

# Charger le Geopackage et lister les couches
geopackage_path = "geoinf-proj-ze-9933.gpkg"
layers = pyogrio.list_layers(geopackage_path)
print("Couches disponibles dans le Geopackage:", layers)

# Lire la couche de la zone industrielle
industrial_zone_layer = 'ze_9933'  # Remplacez par le nom de votre couche
industrial_zone = gpd.read_file(geopackage_path, layer=industrial_zone_layer)

# Vérifier si la couche est vide
if industrial_zone.empty:
    raise ValueError("La couche est vide ou incorrecte.")

# Créer les zones tampons
buffer_500m = industrial_zone.buffer(500)
buffer_1km = industrial_zone.buffer(1000)
buffer_5km = industrial_zone.buffer(5000)

# Fusionner les zones tampons
buffer_500m_union = unary_union(buffer_500m)
buffer_1km_union = unary_union(buffer_1km)
buffer_5km_union = unary_union(buffer_5km)

# Créer les anneaux
anneau_500m_1km = buffer_1km_union.difference(buffer_500m_union)
anneau_1km_5km = buffer_5km_union.difference(buffer_1km_union)

# Créer un GeoDataFrame pour stocker les zones tampons et anneaux
result_gdf = gpd.GeoDataFrame(
    {
        "name": ["500m", "500m-1km", "1km-5km"],
        "geometry": [buffer_500m_union, anneau_500m_1km, anneau_1km_5km],
    },
    crs=industrial_zone.crs,
)

# Sauvegarder les résultats dans un fichier GeoJSON
geojson_path = "zones_tampon.geojson"
result_gdf.to_file(geojson_path, driver="GeoJSON")
print(f"Zones tampons sauvegardées dans le fichier GeoJSON : {geojson_path}")

# Visualisation des résultats
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Afficher la zone industrielle
industrial_zone.plot(ax=ax, color="blue", alpha=0.5, edgecolor="black", label="Zone Industrielle")

# Afficher les tampons et anneaux
gpd.GeoSeries(buffer_500m_union).plot(ax=ax, color="green", alpha=0.4, edgecolor="black", label="500m")
gpd.GeoSeries(anneau_500m_1km).plot(ax=ax, color="orange", alpha=0.4, edgecolor="black", label="500m-1km")
gpd.GeoSeries(anneau_1km_5km).plot(ax=ax, color="red", alpha=0.4, edgecolor="black", label="1km-5km")

# Ajouter une légende et des titres
plt.legend(loc="upper right")
plt.title("Visualisation des zones tampons et anneaux")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Afficher le graphique
plt.show()
```

```python
# Ce script permet de charger une zone industrielle à partir d'un fichier Geopackage,
# ainsi que des couches de communes pour la Suisse. Ensuite, il génère des zones tampons
# autour de la zone industrielle à différentes distances (500m, 1km, 5km), et calcule
# les anneaux entre ces tampons (par exemple, entre 500m et 1km, et entre 1km et 5km).
# Chaque zone tampon est ensuite intersectée avec les communes pour obtenir les
# zones communes entre les tampons et les communes.
# Les résultats sont visualisés sur une carte, et chaque intersection est enregistrée
# dans un fichier GeoPackage avec une couche distincte pour chaque zone tampon.

import geopandas as gpd
import pyogrio
from shapely.ops import unary_union
import matplotlib.pyplot as plt

# Charger le Geopackage et lister les couches
geopackage_path = "geoinf-proj-ze-9933.gpkg"
layers = pyogrio.list_layers(geopackage_path)
print("Couches disponibles dans le Geopackage:", layers)

# Lire la couche de la zone industrielle
industrial_zone_layer = 'ze_9933'  # Remplacez par le nom de votre couche
industrial_zone = gpd.read_file(geopackage_path, layer=industrial_zone_layer)

# Lire la couche des communes pour toute la Suisse
communes_layer = 'communes'  # Remplacez par le nom de la couche des communes
communes = gpd.read_file('admin.gpkg', layer=communes_layer)

# Vérifier si les couches sont vides
if industrial_zone.empty:
    raise ValueError("La couche des zones industrielles est vide ou incorrecte.")
if communes.empty:
    raise ValueError("La couche des communes est vide ou incorrecte.")

# Créer les zones tampons
buffer_500m = industrial_zone.buffer(500)
buffer_1km = industrial_zone.buffer(1000)
buffer_5km = industrial_zone.buffer(5000)

# Fusionner les zones tampons
buffer_500m_union = unary_union(buffer_500m)
buffer_1km_union = unary_union(buffer_1km)
buffer_5km_union = unary_union(buffer_5km)

# Créer les anneaux
anneau_500m_1km = buffer_1km_union.difference(buffer_500m_union)
anneau_1km_5km = buffer_5km_union.difference(buffer_1km_union)

# Préparer les zones tampons dans un dictionnaire
zi_buf = {
    '500m': buffer_500m_union,
    '1km': anneau_500m_1km,
    '5km': anneau_1km_5km
}

# Transformer les zones tampons en GeoDataFrames
zi_buf_gdf = {
    k: gpd.GeoDataFrame(geometry=[poly], crs=industrial_zone.crs)
    for k, poly in zi_buf.items()
}

# Effectuer l'intersection des zones tampons avec les communes
zi_buf_cmnes = {
    k: gpd.overlay(gdf, communes, how='intersection')
    for k, gdf in zi_buf_gdf.items()
}

# Inspecter les résultats pour 500m
print("Zone tampon 500m avant intersection:\n", zi_buf_gdf['500m'])
print("Zone tampon 500m après intersection avec les communes:\n", zi_buf_cmnes['500m'])

# Visualisation des trois zones tampons après intersection
fig, axs = plt.subplots(1, 3, figsize=(20, 10))

for idx, (zone, ax) in enumerate(zip(['500m', '1km', '5km'], axs)):
    zi_buf_cmnes[zone].plot(ax=ax, alpha=0.5, edgecolor="black", label=f"{zone} Intersecté")
    ax.set_title(f"Zone tampon {zone} après intersection avec les communes")
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    ax.legend()

plt.tight_layout()
plt.show()

# Enregistrer les résultats dans un fichier GeoPackage
output_gpkg_path = "resultats_intersection.gpkg"

# Ajouter chaque GeoDataFrame comme une couche distincte
for zone, gdf in zi_buf_cmnes.items():
    gdf.to_file(output_gpkg_path, layer=zone, driver="GPKG", mode="a")

print(f"Les résultats des intersections ont été enregistrés dans {output_gpkg_path}")
```

```python
# Ce script effectue une jointure spatiale entre deux couches géospatiales de points (habitants et emplois)
# et une couche de polygones (par exemple, des zones géographiques ou des communes).
# L'objectif est de calculer la somme d'une colonne (par exemple, la population ou le nombre d'emplois)
# pour chaque polygone en associant les points qui se trouvent à l'intérieur de ces polygones.
# Une nouvelle colonne est ajoutée à la couche des polygones pour stocker les résultats de la somme.
# Le script charge les données à partir d'un fichier Geopackage, vérifie les colonnes disponibles,
# puis applique la jointure spatiale et calcule les sommes pour les différentes zones.

import geopandas as gpd
import pandas as pd

# Fonction pour effectuer la jointure spatiale et calculer les sommes
def sjoin_summarize(poly_lyr, pts_lyr, pts_colname, poly_newcolname):
    """
    poly_lyr:         la couche des polygones
    pts_lyr:          la couche des points
    pts_colname:      la colonne sur laquelle il faut faire la somme
    poly_newcolname:  le nom de la colonne à créer dans la couche
                      des polygones
    """
    # Faire une jointure spatiale pour associer chaque point à un polygone
    sjoin = pts_lyr.sjoin(poly_lyr, how='inner', predicate='within')

    # Group by selon l'index des polygones et calculer la somme
    somme_par_poly = sjoin.groupby(sjoin.index_right)[pts_colname].sum()

    # Enregistrer le résultat dans la couche des polygones
    poly_lyr[poly_newcolname] = poly_lyr.index.map(somme_par_poly).fillna(0)

# Charger les données de points pour les habitants et les emplois
points_path = "ofs_geostat.gpkg"
points_habitants_layer = "pop_2023"  # Remplacez par le nom correct de la couche
points_emplois_layer = "empl_2022"       # Remplacez par le nom correct de la couche

# Lire les couches de points
try:
    points_habitants = gpd.read_file(points_path, layer=points_habitants_layer)
    points_emplois = gpd.read_file(points_path, layer=points_emplois_layer)
except Exception as e:
    raise FileNotFoundError(f"Erreur lors de la lecture des couches : {e}")

# Vérification des colonnes disponibles
print("Colonnes dans points_habitants:", points_habitants.columns)
print("Colonnes dans points_emplois:", points_emplois.columns)
```

```python
# Ce script effectue une jointure spatiale entre des couches de polygones (zones tampons) et des couches de points
# (représentant les habitants et les emplois) pour calculer la somme des habitants et des emplois dans chaque zone tampon.
# Les données sont extraites à partir d'un fichier Geopackage pour les habitants et les emplois,
# puis la fonction `sjoin_summarize` effectue la jointure spatiale pour associer chaque point (habitant ou emploi) à un polygone.
# Le script vérifie l'existence des colonnes nécessaires et assure que les systèmes de référence spatiale (CRS) des données sont cohérents.
# Ensuite, il calcule les sommes pour chaque zone tampon (500m, 1km, 5km), ajoute ces résultats dans les zones respectives,
# et combine les résultats pour générer un tableau avec les informations des habitants et des emplois par zone tampon.
# Enfin, il affiche le tableau final avec les résultats des calculs.

import geopandas as gpd
import pandas as pd

# Fonction pour effectuer la jointure spatiale et calculer les sommes
def sjoin_summarize(poly_lyr, pts_lyr, pts_colname, poly_newcolname):
    """
    poly_lyr:         la couche des polygones
    pts_lyr:          la couche des points
    pts_colname:      la colonne sur laquelle il faut faire la somme
    poly_newcolname:  le nom de la colonne à créer dans la couche
                      des polygones
    """
    # Faire une jointure spatiale pour associer chaque point à un polygone
    sjoin = pts_lyr.sjoin(poly_lyr, how='inner', predicate='within')

    # Faire une opération «group by» selon l'index du polygone, et calculer la somme
    somme_par_poly = sjoin.groupby(sjoin.index_right)[pts_colname].sum()

    # Enregistrer le résultat dans la couche des polygones
    poly_lyr[poly_newcolname] = poly_lyr.index.map(somme_par_poly).fillna(0)

# Charger les données de points pour les habitants et les emplois
points_path = "ofs_geostat.gpkg"
points_habitants_layer = "pop_2023"  # Nom de la couche pour habitants
points_emplois_layer = "empl_2022"       # Nom de la couche pour emplois

# Charger les couches
points_habitants = gpd.read_file(points_path, layer=points_habitants_layer)
points_emplois = gpd.read_file(points_path, layer=points_emplois_layer)

# Vérifier que les colonnes nécessaires existent
if "BBTOT" not in points_habitants.columns:
    raise KeyError("La colonne 'BBTOT' est introuvable dans points_habitants.")
if "B08EMPT" not in points_emplois.columns:
    raise KeyError("La colonne 'B08EMPT' est introuvable dans points_emplois.")

# Vérifier et harmoniser les CRS
for zone_name, zone_gdf in zi_buf_cmnes.items():
    if points_habitants.crs != zone_gdf.crs:
        points_habitants = points_habitants.to_crs(zone_gdf.crs)
    if points_emplois.crs != zone_gdf.crs:
        points_emplois = points_emplois.to_crs(zone_gdf.crs)

# Calculer le nombre d'habitants et d'emplois pour chaque zone tampon
for zone, gdf in zi_buf_cmnes.items():
    # Calcul des habitants
    sjoin_summarize(gdf, points_habitants, pts_colname="BBTOT", poly_newcolname="pop_2023_sum")

    # Calcul des emplois
    sjoin_summarize(gdf, points_emplois, pts_colname="B08EMPT", poly_newcolname="empl_2022_sum")

    # Ajouter une colonne pour identifier le périmètre
    gdf["peri"] = zone

# Extraire les colonnes nécessaires pour chaque zone tampon
try:
    data_500m = zi_buf_cmnes['500m'][['bfs_nummer', 'name', 'peri', 'pop_2023_sum', 'empl_2022_sum']]
    data_1km = zi_buf_cmnes['1km'][['bfs_nummer', 'name', 'peri', 'pop_2023_sum', 'empl_2022_sum']]
    data_5km = zi_buf_cmnes['5km'][['bfs_nummer', 'name', 'peri', 'pop_2023_sum', 'empl_2022_sum']]
except KeyError as e:
    raise KeyError(f"Erreur lors de l'extraction des colonnes pour les zones tampons : {e}")

# Combiner les données dans un seul tableau Pandas
data_combined = pd.concat([data_500m, data_1km, data_5km], ignore_index=True)

# Remplacer les NaN par 0
data_combined = data_combined.fillna(0)

# Afficher le tableau final
print("Tableau final des habitants et emplois par zone tampon et commune :")
print(data_combined)
```

```python
# Ce script traite les résultats des calculs effectués dans l'étape précédente, en effectuant les opérations suivantes :
# 1. Convertir les colonnes "pop_2023_sum" (nombre d'habitants) et "empl_2022_sum" (nombre d'emplois) en types numériques.
# 2. Remplacer les valeurs manquantes (NaN) par 0 dans ces colonnes.
# 3. Calculer plusieurs nouvelles colonnes (Njsem, Nnsem, Njwe, Nnwe) qui représentent des estimations de populations selon différents critères
#    (ex : personnes en emploi et non emploi, avec ou sans emploi, pour différentes périodes de la semaine).
# 4. Exclure les colonnes "pop_2023_sum" et "empl_2022_sum" avant l'affichage et la sauvegarde des résultats.
# 5. Sauvegarder le DataFrame final dans un fichier CSV, à la fois avec les données par périmètre et par commune.
# 6. Générer des fichiers CSV séparés pour les résumés des données par périmètre et par commune.

import pandas as pd
import os

# Définir le chemin du dossier des résultats
resultats_dir = "resultats"
os.makedirs(resultats_dir, exist_ok=True)

# Vérifier et convertir les colonnes en types numériques
data_combined['pop_2023_sum'] = pd.to_numeric(data_combined['pop_2023_sum'], errors='coerce')
data_combined['empl_2022_sum'] = pd.to_numeric(data_combined['empl_2022_sum'], errors='coerce')

# Remplacer les NaN par 0 après conversion
data_combined['pop_2023_sum'] = data_combined['pop_2023_sum'].fillna(0)
data_combined['empl_2022_sum'] = data_combined['empl_2022_sum'].fillna(0)

# Formules pour calculer le nombre de personnes
data_combined['Njsem'] = 0.3 * data_combined['pop_2023_sum'] + 0.8 * data_combined['empl_2022_sum']
data_combined['Nnsem'] = 0.9 * data_combined['pop_2023_sum'] + 0.05 * data_combined['empl_2022_sum']
data_combined['Njwe'] = 0.6 * data_combined['pop_2023_sum'] + 0.05 * data_combined['empl_2022_sum']
data_combined['Nnwe'] = data_combined['pop_2023_sum']

# Exclure les colonnes 'pop_2023_sum' et 'empl_2022_sum' avant affichage et sauvegarde
columns_to_display = ['bfs_nummer', 'name', 'peri', 'Njsem', 'Nnsem', 'Njwe', 'Nnwe']
data_combined_filtered = data_combined[columns_to_display]

# Afficher le DataFrame complet après calculs (sans les colonnes exclues)
print("\n--- DataFrame Complet (data_combined_filtered) ---\n")
print(data_combined_filtered.head())

# Sauvegarder le DataFrame complet dans un fichier CSV
output_path_full = os.path.join(resultats_dir, "personnes_par_peri_et_cmne.csv")
data_combined_filtered.to_csv(output_path_full, index=False, encoding='utf-8')
print(f"Fichier sauvegardé : {output_path_full}")

# Extraction par périmètre
data_peri = data_combined[['peri', 'Njsem', 'Nnsem', 'Njwe', 'Nnwe']]
data_peri_grouped = data_peri.groupby('peri', as_index=False).sum()

# Afficher le DataFrame groupé par périmètre
print("\n--- DataFrame Groupé par Périmètre (data_peri_grouped) ---\n")
print(data_peri_grouped)

# Sauvegarder les données par périmètre dans un fichier CSV
output_path_peri = os.path.join(resultats_dir, "personnes_par_peri.csv")
data_peri_grouped.to_csv(output_path_peri, index=False, encoding='utf-8')
print(f"Fichier sauvegardé : {output_path_peri}")

# Extraction par commune
data_cmne = data_combined[['bfs_nummer', 'name', 'Njsem', 'Nnsem', 'Njwe', 'Nnwe']]
data_cmne_grouped = data_cmne.groupby(['bfs_nummer', 'name'], as_index=False).sum()

# Afficher le DataFrame groupé par commune
print("\n--- DataFrame Groupé par Commune (data_cmne_grouped) ---\n")
print(data_cmne_grouped.head())

# Sauvegarder les données par commune dans un fichier CSV
output_path_cmne = os.path.join(resultats_dir, "personnes_par_cmne.csv")
data_cmne_grouped.to_csv(output_path_cmne, index=False, encoding='utf-8')
print(f"Fichier sauvegardé : {output_path_cmne}")
```
